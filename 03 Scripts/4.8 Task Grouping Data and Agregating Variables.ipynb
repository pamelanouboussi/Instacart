## 03 Step 4: Follow instructions to create loyalty flag

### Aggregating Data with transform()

```python
#Group data by user id 
#create Max order column to agregate results
#creates new max order column, df, grouped by user id, transform function applied to order_num with np max fuction
#np max(assigned max function from the np fibrary)
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#Do not assign options regarding the maximum rows to display(print all rows requested)
pd.options.display.max_rows = None
```

```python
ords_prods_merge.head(100)
```

### Deriving Columns with loc()

```python
#Create flags to assign loyalty label to user id based on corresponding max order value
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#View frequency of nnew loyaly flag column
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

## 04 Step 4 :Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer).

```python
# Split the data into groups by "loyalty_flag" and calculate the mean, max, min for each "prices" column
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'max', 'min']})
```

## 05 Step 6: Create Spending Flag
Create a spending flag for each user based on the average price across all their orders using the following criteria:
If the mean of the prices of products purchased by a user is lower than 10, then flag them as a Low spender.
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a High spender.

```python
#Group data by user id and find mean price for each user. create mean price column for results
ords_prods_merge['mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#error check
ords_prods_merge.head(100)
```

```python
#Create flag to assign spending label
ords_prods_merge.loc[ords_prods_merge['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
#Frequancy check
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

## 06 Step 7: Create Order frequency flag

Create an order frequency flag that marks the regularity of a users ordering behavior according to the median in the days_since_prior_order column. The criteria for the flag should be as follows:
If the median of days_since_prior_order is higher than 20, then the customer should be labeled a Non-frequent customer.
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a Regular customer.
If the median is lower than or equal to 10, then the customer should be labeled a Frequent customer.

```python
#Group data by user id column. create new colum to aggregate results
ords_prods_merge['median_days_since_last_order'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
#Check results
ords_prods_merge.head(100)
```

```python
#Create flag to assign frequency label to user id, with median days since last order
ords_prods_merge.loc[ords_prods_merge['median_days_since_last_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python

ords_prods_merge.loc[(ords_prods_merge['median_days_since_last_order'] > 10) & (ords_prods_merge['median_days_since_last_order'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python

ords_prods_merge.loc[ords_prods_merge['median_days_since_last_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
#Frequency check
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```

```python
#looking at the null values 
print(ords_prods_merge[ords_prods_merge.frequency_flag.isnull()])
```

```python
# Customers with one order are given NAN values.
#Change null values in frequency flag column to Non-frequent customer
ords_prods_merge['frequency_flag'].fillna("Non-frequent customer", inplace = True)
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```
