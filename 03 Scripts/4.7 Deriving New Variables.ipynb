## 01 Creating a user defined function

```python
#Create subset of df
df = ords_prods_merge[:1000000]
```

```python
#Create function: assign each row with label low,mid,high based on its price
#define function with the def syntax, give name to new function (price_label). 
#Use the row argument to look at at every row.
#elif (else if) function adds aditional conditions to if statement
#complete with a colon. it separates the head (name and arguments) and body (what it does)of the function.
# Used indentation to show lines of code belong within the code that comes before it. 

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Apply the function
#axis 1 refers to rows, the syntax df() is used to create new collum within df to place new labels
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna=False)
```

```python
#check the most expensive product in df subset
df['prices'].max()
```

## 02 If-Statements with the loc() Function

```python
#call loc function on df. if price is greater than 15, set as high range product
#the comma separates the implied if from the then
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
#call loc function on df. if price is greater than 5,but less than or equal to 15 set as mid range product
#conditions combined with the & sign. both place in brackets to be treadted as separate conditions
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
#call loc function on df. if price is less than 5, set as low range product
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#count the number of products within each label
df['price_range_loc'].value_counts(dropna=False)
```

### If-Statements with For-Loops

```python
# find out which days most orders take place with a frequency
ords_prods_merge['orders_day_of_week'].value_counts(dropna=False)
```

```python
# Run for loop code: create an empty list called result (to place results) 
#for each value, something needs to be one
#if values in row are 0, apply busy day, if 4 least busy if none, regularly busy (value acts as placeholder)

result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Create a new collum in df to combine results with df
ords_prods_merge['busiest_day']= result
```

```python
#count the number of busy days within each label
ords_prods_merge['busiest_day'].value_counts(dropna=False)
```

### Identify busiest hours. Label periods of time  “Most orders,” “Average orders,” and “Fewest orders.” 
### Create a new column containing these labels called “busiest_period_of_day.”

```python
#Create frequency collum of order hour of day
ords_prods_merge['order_hour_of_day'].value_counts()
```

#### Observations
The busiest hours are 10,11,14,15,13,12,16,9. the Average orders hours are 17,8,19,20,7,21,22, 23 and the least busy hours are 6,0,1,5,2,4,3.

```python
# Create a for-loop code usind above frequency table
hours_of_day = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value in [10,11,14,15,13,12,16,9]:
    hours_of_day.append("Most orders")
  elif value in [6,0,1,5,2,4,3]:
    hours_of_day.append("Fewest orders")
  else:
    hours_of_day.append("Average orders")
```

```python
# Create a new collum in df to combine results with df
ords_prods_merge['Busiest_period_of_day']= hours_of_day
```
