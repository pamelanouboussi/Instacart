## 02 Aggregating Data with agg()


#### Performing Single Aggregations

```python
#Find the mean order number and group by department id
#assign group by function to df. apply the agg function to the results
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
#perform command above with the mean function

df.groupby('department_id')['order_number'].mean()
```

NB: When using mean() (or any other standard aggregation function), simply index the column with square brackets, followed by function
-Column names can also be set off with dot :df.groupby('department_id').order_number.mean()

#### Performing Multiple Aggregations

```python
#Calculate mean, max and min
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

### Aggregating Data with transform()


```python
#create Max order column
#creates new max order column, df, grouped by user id, transform function applied to order_num with np max fuction
#np max(assigned max function from the np fibrary)
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

### Deriving Columns with loc()

```python
#Create flags to assign loyalty label to user id based on corresponding max order value
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#View frequency of nnew loyaly flag column
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```
